```python
# Quartiles
df.quantile([0.25, 0.50, 0.75], numeric_only=True)
```

```python
# Interquartile Range (IQR)
Q1 = df['balance'].quantile(0.25)
Q3 = df['balance'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# Upper Limit Outliner limit
upper_limit = Q3 + (1.5 * IQR)

upper_limit
```

```python
# Lower Limit  Outliner limit
lower_limit = Q1 - (1.5 * IQR)

lower_limit
```

```python
# Identifying Outliers
outliers = df[(df['balance'] < lower_limit) | (df['balance'] > upper_limit)]
```

```python
# Number of Outliers in dataset
len(outliers)
```

```python
# Percentiles
df.quantile([0.10,0.25,0.50,0.75,0.90], numeric_only=True)
```

```python
# Kurtosis
df.kurt(numeric_only=True)
```

## EDA Observations

Based on the exploratory data analysis, the following observations were made:

### Measures of Central Tendency

* The average customer age is 40.94 years, while the median age is 39 years, indicating a relatively balanced age distribution.
* The average account balance is 1,362.27, whereas the median balance is 448, suggesting the presence of customers with exceptionally high balances.
* The average call duration is 258.16 seconds, while the median duration is 180 seconds, indicating that some calls lasted significantly longer than most calls.

### Measures of Dispersion

* The Balance variable shows the highest variability with a standard deviation of 3044.77, indicating substantial differences in customer account balances.
* The Duration variable also exhibits high variability with a standard deviation of 257.53.
* The range of Balance values is 110,146, demonstrating a very wide spread in customer balances.

### Quartiles and Percentiles

* 25% of customers have balances below 72.
* 50% of customers have balances below 448.
* 75% of customers have balances below 1,428.
* 90% of customers have balances below 3,574.

### Outlier Analysis

* The Interquartile Range (IQR) for Balance is 1,356.
* The lower outlier limit is -1,962.
* The upper outlier limit is 3,462.
* Using the IQR method, 4,729 observations were identified as outliers in the Balance variable.
* These outliers were retained as they may represent genuine customer characteristics rather than data entry errors.

### Skewness

* Balance-8.36, Campaign-4.90, Duration-3.14, Pdays-2.62, and Previous-41.85 exhibit strong positive skewness.
* This indicates that a small number of observations have substantially larger values than the majority of observations.

### Kurtosis

* Balance 140.75 and Previous 4506.86 exhibit extremely high kurtosis values.
* This confirms the presence of extreme observations and heavy tails in these variables.
* Day is the only variable with negative kurtosis (-1.06), indicating a relatively flatter distribution.

### Overall Conclusion

The dataset contains substantial variability and several highly skewed variables, particularly Balance, Duration, Campaign, Pdays, and Previous. Outlier analysis confirms the presence of extreme values, especially in customer account balances. These characteristics should be considered while performing further analysis and interpretation.
